```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data yang disesuaikan dengan hasil terminal asli
data = {
    'Algoritma': ['Logistic Regression', 'Logistic Regression', 'Logistic Regression', 
                  'XGBoost', 'XGBoost', 'XGBoost'],
    'Metrik': ['Akurasi', 'Recall', 'Precision', 'Akurasi', 'Recall', 'Precision'],
    'Skor': [98.75, 98.75, 98.79, 98.75, 98.00, 100.00] # Angka sesuai gambar terminal
}

df = pd.DataFrame(data)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

palette = {"Logistic Regression": "#4CAF50", "XGBoost": "#F26D17"}
ax = sns.barplot(x='Metrik', y='Skor', hue='Algoritma', data=df, palette=palette)

# Menambahkan label angka di atas batang
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.2f}%', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha = 'center', va = 'center', 
                    xytext = (0, 9), 
                    textcoords = 'offset points',
                    fontsize=10, fontweight='bold')

plt.title('Perbandingan Performa: Logistic Regression vs XGBoost', fontsize=16, pad=20)
plt.ylabel('Skor (%)', fontsize=12)
plt.ylim(90, 105) # Range Y yang pas agar terlihat jelas
plt.legend(title='Algoritma', loc='lower right')
plt.tight_layout()
plt.show()
```
